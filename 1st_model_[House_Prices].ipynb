Visualizing data

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True)
plt.title('Distribution of Housing Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show
```

Dropping unneeded features

```python
df_refined = df[['area', 'price']]
print (f"Original: {df.shape}")
print (f"Refined: {df_refined.shape}")
```

Defining Input (X) and Target (Y)

```python
X = df_refined[['area']].values
Y = df_refined[['price']].values
```

Importing and setting params for training

```python
from sklearn.model_selection import train_test_split

# 80% for training, 20% for testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

Importing LR model and training it

```python
from sklearn.linear_model import LinearRegression

# Create a linear regression model
model = LinearRegression()

model.fit(X_train, Y_train)
```

Checking Coefficiency and Interception

```python
print(f"Model Coefficient: {model.coef_[0]}")
print(f"Model Intercept: {model.intercept_}")
```

Testing the model

```python
from sklearn.metrics import mean_absolute_error

y_pred = model.predict(X_test)
mae = mean_absolute_error(Y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
```

```python
from sklearn.metrics import r2_score

# Calculate R-squared
r2 = r2_score(Y_test, y_pred)
print(f"R-squared score: {r2}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, y_pred, alpha=0.3, color='blue')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'k--', lw=3)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs. Predicted Housing Prices')
plt.show()
```

Checking if overfit

```python
# Score on training data
train_score = model.score(X_train, Y_train)

# Score on test data
test_score = model.score(X_test, Y_test)

print(f"Training R2: {train_score}")
print(f"Test R2: {test_score}")
```
